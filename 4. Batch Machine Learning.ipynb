# 4. Batch Machine Learning

# Code Example (Batch Training)

```python
from sklearn.linear_model import LinearRegression

# Entire dataset
X = [[1], [2], [3], [4], [5]]
y = [10, 20, 30, 40, 50]

# Train model on full batch
model = LinearRegression()
model.fit(X, y)

# Prediction
print(model.predict([[6]]))

# Model trained once on full dataset
```
