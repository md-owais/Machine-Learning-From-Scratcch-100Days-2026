```python
We’ll focus on:

🔥 Semi-Supervised Learning
🚀 Self-Supervised Learning

Both sit between traditional supervised and unsupervised learning in Artificial Intelligence.

🔥 1. Semi-Supervised Learning
✅ Definition

Uses a small amount of labeled data + a large amount of unlabeled data.

👉 Real-world reason:
Labeling data is expensive, but unlabeled data is abundant.

🧩 How it Works
Train model on small labeled dataset
Use it to predict labels for unlabeled data
Retrain using both (pseudo-labeling)
📊 Techniques
🔹 1. Self-training (Pseudo-labeling)
Model generates its own labels
🔹 2. Label Propagation
Labels spread based on similarity
🔹 3. Co-training
Two models train each other
🌍 Real-World Examples
Medical imaging (few labeled scans, many unlabeled)
Web data classification
Speech recognition

👉 Used in large-scale systems like Google where labeling billions of data points is impossible

💻 Code Example (Pseudo-labeling Concept)
from sklearn.ensemble import RandomForestClassifier
import numpy as np

# Labeled data
X_labeled = np.array([[1], [2], [3]])
y_labeled = np.array([0, 0, 1])

# Unlabeled data
X_unlabeled = np.array([[4], [5]])

# Train initial model
model = RandomForestClassifier()
model.fit(X_labeled, y_labeled)

# Predict pseudo labels
pseudo_labels = model.predict(X_unlabeled)

# Combine data
X_combined = np.vstack((X_labeled, X_unlabeled))
y_combined = np.hstack((y_labeled, pseudo_labels))

# Retrain
model.fit(X_combined, y_combined)
📌 Key Points
Reduces labeling cost
Improves performance with limited labels
Risk: wrong pseudo-labels can propagate errors
🚀 2. Self-Supervised Learning (🔥 Very Important for GenAI)
✅ Definition

Model learns by creating labels from the data itself (no manual labeling).

👉 This is the backbone of modern AI like
ChatGPT and LLMs

🧩 How it Works

Instead of labels:

Create pretext tasks
Examples:
Predict missing word
Predict next sentence
Fill missing part of image
📊 Common Techniques
🔹 1. NLP (Text)
Masked Language Modeling (MLM)
Next Token Prediction

👉 Used in models like BERT

🔹 2. Computer Vision
Image rotation prediction
Contrastive learning
🔹 3. Contrastive Learning (🔥 Important)
Learn by comparing similar vs dissimilar samples
🌍 Real-World Examples
Language models (chatbots, translation)
Image understanding
Speech models

👉 Used by companies like OpenAI

💻 Code Example (Conceptual NLP)
# Self-supervised example (masked word prediction)

sentence = "Machine learning is ___"

# Model tries to predict missing word
# Expected output: "powerful"

# In real-world, transformers handle this automatically

👉 Real implementation uses transformers (Hugging Face, etc.)

📌 Key Points
No manual labeling required
Scales to massive datasets
Foundation of LLMs & GenAI
⚔️ Semi vs Self-Supervised (Important)
Feature	Semi-Supervised	Self-Supervised
Labels	Few real labels	No real labels
Data	Labeled + Unlabeled	Only unlabeled
Approach	Use pseudo-labels	Create tasks
Use Case	Limited labeled data	Massive data (LLMs)
Risk	Label noise	Task design matters
🧠 Simple Analogy
Semi-supervised → Teacher gives few answers, student guesses rest
Self-supervised → Student creates own questions and solves them
🎯 Interview-Ready Answer

👉 “Semi-supervised learning uses a combination of small labeled data and large unlabeled data to improve learning efficiency, while self-supervised learning generates supervisory signals from the data itself using pretext tasks, and is widely used in modern deep learning models like LLMs.”

🚀 Pro Tip (For YOU – Very Important)

Since you’re targeting GenAI / Data Scientist roles:

Self-supervised = 🔥 MUST KNOW (LLMs, transformers)
Semi-supervised = ⭐ Bonus + strong practical knowledge
Mention contrastive learning + transformers → big plus
```
